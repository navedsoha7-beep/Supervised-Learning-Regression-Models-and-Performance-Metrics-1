1.  What is Simple Linear Regression (SLR)? Explain its purpose.


    - - Simple Linear Regression (SLR) is a statistical method used to model the relationship between two variables:
- Independent variable (X) → the predictor or input
- Dependent variable (Y) → the outcome or target
- It assumes that the relationship between X and Y can be described by a straight line equation:
Y=\beta _0+\beta _1X+\epsilon




2. What are the key assumptions of Simple Linear Regression?


     - The key assumptions of Simple Linear Regression are: linearity between variables, independence of residuals, constant variance of errors (homoscedasticity), and normal distribution of residuals. These assumptions ensure the regression model is valid, interpretable, and statistically reliable.”





3. Write the mathematical equation for a simple linear regression model and
explain each term.




     - The equation of a simple linear regression model is Y=\beta _0+\beta _1X+\epsilon . Here, Y is the dependent variable, X is the independent variable, \beta _0 is the intercept, \beta _1 is the slope showing the effect of X on Y, and \epsilon  is the error term accounting for unexplained variation.

5. What is the method of least squares in linear regression?



      - he method of least squares is the most common technique used to estimate the parameters (\beta _0,\beta _1) in a linear regression model.
- It works by minimizing the sum of the squared differences between the actual values (Y) and the predicted values (\hat {Y}) from the regression line.
Mathematically, it minimizes:
\mathrm{SSE}=\sum _{i=1}^n(Y_i-\hat {Y}_i)^2

8.  What is the purpose of the R-squared metric in regression analysis?



     -  R-squared (Coefficient of Determination) measures how well the regression model explains the variability of the dependent variable (Y).






    



```python
#9.Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

from sklearn.linear_model import LinearRegression



X = [[1], [2], [3], [4], [5]]   # Independent variable (study hours)
y = [20, 40, 60, 80, 100]       # Dependent variable (exam marks)



model=LinearRegression()
model.fit(X,y)


print("Slope (β1):", model.coef_[0])
print("Intercept (β0):", model.intercept_)

# Predict marks for 9 hours of study
predicted = model.predict([[9]])
print("Predicted marks for 9 hours study:", predicted[0])


```

10. How do you interpret the coefficients in a simple linear regression model?




     - In a simple linear regression model, the intercept represents the predicted value of the dependent variable when the independent variable is zero. The slope represents how much the dependent variable changes for each one-unit increase in the independent variable. A positive slope indicates a direct relationship, while a negative slope indicates an inverse relationship.
